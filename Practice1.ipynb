```python
url = "https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv"
baby = pd.read_csv(url)
baby
```

```python
boys = baby[baby["Gender"]=="MALE"]
girls = baby[baby["Gender"]=="FEMALE"]
```

```python
boys.sort_values("Rank")
```

```python
excel_file = pd.ExcelWriter("baby_names.xlsx")
```

```python
girls.to_excel(excel_file, sheet_name="Girls", index= False)
boys.to_excel(excel_file, sheet_name="Boys", index= False)
```

```python
excel_file.save()
```

```python
from pandas_datareader import data
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bb = data.DataReader(name="BB", data_source="yahoo", start="2007-01-01", end="2022-02-22")
bb.head(3)
```

```python
bb.plot(y=["Adj Close"], kind="line")
```

```python
bb[["High", "Low"]].plot(kind="line")
```

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
plt.style.use('dark_background')
plt.style.use('seaborn-whitegrid')
```

```python
bb.plot(y = "High")
```

```python
def rank_performance(stock_price):
    if stock_price <= 10:
        return "poor"
    elif stock_price <= 50:
        return "Satisfactory"
    else:
        return "Steller"
    
bb["Close"].apply(rank_performance).value_counts()
```

```python
plt.style.use("ggplot")
bb["Close"].apply(rank_performance).value_counts().plot(kind="barh")
```

```python
bb["Close"].apply(rank_performance).value_counts().plot(kind="bar")
```

```python
bb["Close"].apply(rank_performance).value_counts().plot(kind="pie", legend=True)
```
